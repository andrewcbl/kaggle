```python
import graphlab
```

# Read in the train and test data

```python
train = graphlab.SFrame("train.csv")
test = graphlab.SFrame("test.csv")
```

# Data cleaning

## Fill the age with mean value

```python
train = train.fillna('Age', train['Age'].mean())
```

```python
train.add_column(train['Sex'] == 'female', 'SexInt')
```

```python
train.head()
```

```python
test = test.fillna('Age', train['Age'].mean())
test.add_column(test['Sex'] == 'female', 'SexInt')
```

```python
train_train, train_test = train.random_split(0.75, seed=0)
```

```python
len(train_train)
```

```python
model_logistic = graphlab.logistic_classifier.create(train_train, target='Survived', features=['Age', 'SexInt'])
```

```python
model_logistic.evaluate(train_test)
```

```python
test_pred = model_logistic.predict(test)
```

```python
submission = graphlab.SFrame({"PassengerId": test['PassengerId'], "Survived": test_pred})
```

```python
submission.save("kaggle.csv", format='csv')
```
